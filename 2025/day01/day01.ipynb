## Part 1


```python
from collections import namedtuple

Rotation = namedtuple("Rotation", "d n")

def parse_data(s):
    return [
        Rotation(line[0], int(line[1:]))
        for line in s.strip().splitlines()
    ]

def part_1(data):
    d = {"L": -1, "R": 1}
    location = 50
    hits_zero = 0
    for r in data:
        location = (location + d[r.d] * r.n) % 100
        if location == 0:
            hits_zero += 1
    return hits_zero
    
test_data = parse_data("""L68
L30
R48
L5
R60
L55
L1
L99
R14
L82""")

assert part_1(test_data) == 3
```

```python
data = parse_data(open("input").read())
part_1(data)
```

```python
def part_2(data):
    d = {"L": -1, "R": 1}
    location = 50
    hits_zero = 0
    for r in data:
        for _ in range(r.n):
            location = (location + d[r.d]) % 100
            if location == 0:
                hits_zero += 1
    return hits_zero

assert part_2(test_data) == 6
```
