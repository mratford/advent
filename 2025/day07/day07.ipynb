## Part 1


```python
import advent

def parse_data(s):
    rows = s.strip().splitlines()
    return (
        rows[0].find("S"), 
        [
            {x for x, c in enumerate(row) if c == "^"}
            for row in rows
        ]
    )
    
def part_1(data):
    p, splitter_map = data
    ps = {p}
    n_splits = 0
    for splitters in splitter_map:
        splits = ps & splitters
        n_splits += len(splits)
        ps -= splits
        for d in (-1, 1):
            ps |= {x + d for x in splits}
    return n_splits

test_data = parse_data(""".......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
""")

assert part_1(test_data) == 21
```

```python
data = parse_data(advent.read_input())

part_1(data)
```

```python
from functools import cache

def part_2(data):
    p, splitter_map = data

    @cache
    def timelines(p, row):
        if row == len(splitter_map):
            return 1
        elif p in splitter_map[row]:
            return timelines(p-1, row+1) + timelines(p+1, row+1) 
        else:
            return timelines(p, row+1)

    return timelines(p,0)
    

assert part_2(test_data) == 40
```
