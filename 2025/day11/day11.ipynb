## Part 1
This should be OK with memoised recursion? I'm nervous after yesterday.

```python
from advent import read_input

def parse_data(s):
    g = {}
    for line in s.strip().splitlines():
        fields = line.strip().split()
        g[fields[0][:-1]] = fields[1:]
    return g

test_data = parse_data("""aaa: you hhh
you: bbb ccc
bbb: ddd eee
ccc: ddd eee fff
ddd: ggg
eee: out
fff: out
ggg: out
hhh: ccc fff iii
iii: out
""")

test_data
```

```python
from functools import cache

def n_paths(data):
    @cache
    def n_paths_from(x):
        if x == "out":
            return 1
        else:
            return sum(n_paths_from(nbr) for nbr in data.get(x, []))

    return n_paths_from("you")
    
assert n_paths(test_data) == 5
```

```python
data = parse_data(read_input())

n_paths(data)
```

## Part 2

A tweak to test the validity of the path will hopefully still be ok.

```python
def n_paths_with_fft_and_dac(data):
    @cache
    def n_paths_from(x, dac, fft):
        if x == "out":
            return 1 if dac and fft else 0
        else:
            return sum(
                n_paths_from(
                    nbr,
                    dac or x == "dac",
                    fft or x == "fft"
                )
                for nbr in data.get(x, [])
            )

    return n_paths_from("svr", False, False)

test_data_2 = parse_data("""svr: aaa bbb
aaa: fft
fft: ccc
bbb: tty
tty: ccc
ccc: ddd eee
ddd: hub
hub: fff
eee: dac
dac: fff
fff: ggg hhh
ggg: out
hhh: out""")

assert n_paths_with_fft_and_dac(test_data_2) == 2
```

```python
n_paths_with_fft_and_dac(data)
```
