## Part 1


```python
import advent
from math import prod

def parse_data(s):
    return [
        [int(x) for x in column[:-1]] + [column[-1]]
        for column in list(
            zip(
                *[
                    line.strip().split() 
                    for line in s.strip().splitlines()
                ]
            )
        )
    ]
    
test_data = parse_data("""123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
""")

test_data
```

```python
def part_1(data):
    ops = { "*": prod, "+": sum }
    return sum(
        ops[row[-1]](row[:-1])
        for row in data
    )

assert part_1(test_data) == 4277556
```

```python
data = parse_data(advent.read_input())

part_1(data)
```

## Part 2 
This is a bit tedious.

```python
def parse_data_2(s):
    return [line for line in s.splitlines() if line]

def op_indices(row):
    result = []
    for i, x in enumerate(row):
        if x != " ":
            result.append(i)
    return result

def numbers_from_columns(rows, start, end):
    for c in range(start, end):
        result = 0
        for r in rows:
            if r[c] != " ":
                result = result * 10 + int(r[c])
        if result > 0:
            yield result

def part_2(data):
    op_row = data[-1]
    number_rows = data[:-1]
    ops = { "*": prod, "+": sum }
    result = 0
    op_is = op_indices(op_row)
    for start, end in zip(op_is, op_is[1:] + [len(op_row)]):
        result += ops[op_row[start]](numbers_from_columns(number_rows, start, end))
    return result

test_data_2 = parse_data_2("""123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
""")

assert part_2(test_data_2) == 3263827
```

```python
data_2 = parse_data_2(advent.read_input())

part_2(data_2)
```
