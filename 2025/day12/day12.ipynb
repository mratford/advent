## Part 1

I wasn't going to bother with this one until I read that the number of 3x3 blocks that fit into the area was working as a solution.

```python
from advent import read_input

# Quicker to eyeball than parse
min_occupancy = {0: 7, 1: 5, 2: 7, 3: 6, 4: 7, 5: 7}

blocks = read_input().strip().split("\n\n")
lines = blocks[-1].splitlines()
definitely_not_solvable = 0
definitely_solvable = 0
all_count = 0
for line in lines:
    parts = line.split(":")
    cs = parts[0].split("x")
    x = int(cs[0])
    y = int(cs[1])
    naive_blocks_to_fit = (x//3) * (y//3)
    area = x * y
    blocks = [int(x) for x in parts[1].strip().split()]
    if sum(blocks) <= naive_blocks_to_fit:
        definitely_solvable += 1
    if sum(min_occupancy[i] * x for i, x in enumerate(blocks)) > area:
        definitely_not_solvable += 1
    all_count += 1
print(all_count)
print(definitely_solvable)
print(definitely_not_solvable)
```
