## Part 1
A greedy algorithm where the first digit is the maximum available works here.

```python
def parse_data(s):
    return [
        [int(x) for x in line]
        for line in s.strip().splitlines()
    ]

test_data = parse_data("""987654321111111
811111111111119
234234234234278
818181911112111""")

test_data
```

```python
def largest_joltage(bank):
    first_i = max(range(len(bank) - 1), key=lambda i: bank[i])
    second_i = max(range(first_i + 1, len(bank)), key=lambda i: bank[i])
    return bank[first_i] * 10 + bank[second_i]

def part_1(data):
    return sum(largest_joltage(bank) for bank in data)

assert part_1(test_data) == 357
```

```python
data = parse_data(open("input").read())

part_1(data)
```

## Part 2
Extend that, always picking the maximum in the digits left.

```python
def largest_joltage_2(bank):
    result = 0
    last_i = -1
    for i in range(12):
        next_i = max(range(last_i + 1, len(bank) - (11 - i)), key=lambda i: bank[i])
        result = result * 10 + bank[next_i]
        last_i = next_i
    return result

def part_2(data):
    return sum(largest_joltage_2(bank) for bank in data)

assert part_2(test_data) == 3121910778619
```

```python
part_2(data)
```
